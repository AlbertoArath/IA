## Moons

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=(200,200), noise=0.05, random_state=33)

print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='blue')
```

```python
from sklearn.cluster import KMeans

k = 2
kmeans_model = KMeans(n_clusters=k,
                      random_state=23).fit(X)
```

```python
labels = kmeans_model.labels_
labels[:200]
```

```python
centroids = kmeans_model.cluster_centers_
centroids
```

```python

plt.scatter(X[labels == 0, 0], X[labels == 0, 1], c = 'red')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], c = 'blue')
plt.scatter(centroids[0, 0], centroids[0, 1], c = 'red', marker = "X", s = 400)
plt.scatter(centroids[1, 0], centroids[1, 1], c = 'blue', marker = "X", s = 400)



```
