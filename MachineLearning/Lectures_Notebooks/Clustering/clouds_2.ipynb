## Clouds

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=200, 
                           n_features=2, 
                           n_informative=2,
                           n_redundant=0, 
                           n_classes=4,
                           class_sep=2.5, 
                           n_clusters_per_class=1,
                           random_state=10)

print(X.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='blue')
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering


k = 2
model = AgglomerativeClustering(n_clusters=2,
                      linkage = 'average').fit(X)
```

```python
#centroids = model.cluster_centers_
labels = model.labels_
#centroids
labels
```

```python

plt.scatter(X[labels == 0, 0], X[labels == 0, 1], c = 'red')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], c = 'blue')
#plt.scatter(centroids[0, 0], centroids[0, 1], c = 'red', marker = "X", s = 400)
#plt.scatter(centroids[1, 0], centroids[1, 1], c = 'blue', marker = "X", s = 400)

```
