```python
import kagglehub
import pandas as pd
from pathlib import Path

# Download latest version
path = Path (kagglehub.dataset_download("yasserh/titanic-dataset"))

print ("Path to dataset files:", path)

data_set_name = Path ("Titanic-Dataset.csv")

df = pd.read_csv(path / data_set_name, header=None)

df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
unique_cols = [col for col in df.columns if df[col].nunique() == df.shape[0]]
print("Variables con valores Ãºnicos (poco Ãºtiles para clasificaciÃ³n):", unique_cols)
```

Variables categÃ³ricas y numÃ©ricas

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Variables categÃ³ricas:", categorical_cols)
print("Variables numÃ©ricas:", numeric_cols)
```

```python
# Calculamos la varianza de las variables numÃ©ricas
variances = df[numeric_cols].var()
variances_sorted = variances.sort_values(ascending=False)
print("Varianzas de variables numÃ©ricas:\n", variances_sorted)

# Seleccionamos las dos variables con mayor varianza
vars_alta_var = variances_sorted.index[:2].tolist()
print("Variables con mayor varianza:", vars_alta_var)

# Graficamos histogramas para visualizar su distribuciÃ³n
for col in vars_alta_var:
    plt.figure()
    plt.hist(df[col].dropna(), bins=30, alpha=0.7)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
vars_baja_var = variances_sorted.index[-2:].tolist()
print("Variables con menor varianza:", vars_baja_var)

# Graficamos sus histogramas
for col in vars_baja_var:
    plt.figure()
    plt.hist(df[col].dropna(), bins=range(int(df[col].min()), int(df[col].max())+2), alpha=0.7, align='left')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Agrupamos los supervivientes por sexo
survivors_by_sex = df[df['Survived'] == 1]['Sex'].value_counts()
print("Supervivientes por sexo:")
print(survivors_by_sex)

# GrÃ¡fico de barras
plt.figure()
survivors_by_sex.plot(kind='bar', title='Supervivientes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```
