- ¿Qué obras se han presentado en más de un teatro?

```python
shows_grouped = short_df.groupby("Show.Name")["Show.Theatre"].unique()
shows_grouped_total_counts = short_df.groupby("Show.Name")["Show.Theatre"].nunique()

print(shows_grouped)
print(shows_grouped_total_counts)
#
# Restore data frame
#
shows_grouped_total_counts = shows_grouped_total_counts.reset_index()
shows_grouped_total_counts[shows_grouped_total_counts ["Show.Theatre"] >= 2]
```
