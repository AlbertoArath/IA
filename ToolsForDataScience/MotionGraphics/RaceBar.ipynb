```python
import matplotlib.pyplot as plt
import numpy    as np
from matplotlib import animation


fig = plt.figure(figsize = (5,3))

ax = fig.add_subplot()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim((-1.1, 1.1))
ax.grid(True)
ax.set_facecolor("xkcd:pistachio")

linea, = ax.plot([],[], lw=2)
n = 20

def init ():
    x = np.linspace (0, 2*np.pi, n)
    y = np.sin(x)
    linea.set_data(x, y)
    pass

f = 10

def animate(i):
    x = np.linspace (0, 2*np.pi, n)
    y = np.sin(x + 2 * np.pi * i / f)
    linea.set_data(x, y)
    return [linea]
```

```python
from IPython.display import HTML

anim = animation.FuncAnimation(fig, animate, init_func = init, frames = f)
HTML (anim.to_jshtml())
```

```python

import matplotlib.pyplot as plt
from matplotlib import animation

fig = plt.figure (figsize = (8, 3))
ax = fig.add_subplot ()

ax.set_xlim (0, 5.5)
ax.set_ylim (.5, 3.5)
ax.set_yticks ([1, 2, 3])

v1 = [1, 3, 5]
abars = ax.barh ([1, 2,3], v1)
n_frames = 30
def animate (i):
    ax.clear()
    
    ax.set_xlim (0, 5.5)
    ax.set_ylim (.5, 3.5)
    ax.set_yticks ([1, 2, 3])

    z = ax.barh([1,2,3], (v1[n_frames % 3], 0, 0))
    return z

anim = animation.FuncAnimation(fig, animate, frames = 3, interval=100)
HTML (anim.to_jshtml())
```
