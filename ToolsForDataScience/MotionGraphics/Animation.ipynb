```python
import matplotlib.pyplot as plt
import numpy    as np
from matplotlib import animation


fig = plt.figure(figsize = (5,3))

ax = fig.add_subplot()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim((-1.1, 1.1))
ax.grid(True)
ax.set_facecolor("xkcd:pistachio")

linea, = ax.plot([],[], lw=2)
n = 100

def init ():
    x = np.linspace (0, 2*np.pi, n)
    y = np.sin(x)
    linea.set_data(x, y)
    pass

f = 20

def animate(i):
    x = np.linspace (0, 2*np.pi, n)
    y = np.sin(x + 2 * np.pi * i / f)
    linea.set_data(x, y)
    return [linea]
```

```python
from IPython.display import HTML

anim = animation.FuncAnimation(fig, animate, init_func = init, frames = f)
HTML (anim.to_jshtml())
```
