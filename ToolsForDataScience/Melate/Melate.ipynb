```python
#print ("Diff", melate_series.diff())
melate_series_diff = melate_series.diff()
melate_series_diff = melate_series_diff.dropna()
print ("Melate series", melate_series_diff[melate_series_diff == 0])
#print ("Diff", melate_series_diff.head(20))
```

```python

prev_game = 0
strike = 0
largest_strike_no_winner = 0
consecutive_win = 0
largest_consecutive_win = 0


melate_series.diff()

for game in melate_series_diff.values:


  if game > prev_game:
    #print ("1")
    strike = strike + 1
    if consecutive_win > largest_consecutive_win:
      largest_consecutive_win = consecutive_win
    consecutive_win = 0
  elif game == prev_game:
    if (strike > largest_strike_no_winner):
      largest_strike_no_winner  = strike   
    consecutive_win += 1
    strike = 0
  else:
    #print ("0")
    if strike > largest_strike_no_winner:
      largest_strike_no_winner  = strike
    if consecutive_win > largest_consecutive_win:
      largest_consecutive_win = consecutive_win
    consecutive_win = 0
    strike = 0
  prev_game = game

#Contesta las siguientes preguntas:
#a. ¿Cuál es la mayor cantidad de sorteos que han pasado sin obtener un ganador?
#a. ¿Cuál es la vez en que ha habido ganadores de melate más próximos?
print (largest_strike_no_winner)
print (largest_consecutive_win)
```
