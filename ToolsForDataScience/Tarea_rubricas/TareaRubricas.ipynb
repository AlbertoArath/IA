```python
with open("caracterizaciones.txt", 'r') as file:
    caracterizaciones = file.readlines()

with open("rubricas.csv", 'r') as file:
    rubricas_content = file.readlines()

caracterizaciones_clean = []
for each_line in caracterizaciones:
    caracterizaciones_clean.append(each_line.replace("\n", "").split()[1:])

rubricas_clean = []
for each_line in rubricas_content:
    rubricas_clean.append(each_line.replace("\n", "").split(";")[2:])

#print (caracterizaciones_clean)

map_rubric_characterization = [dict(zip(x, y)) for x, y in zip (rubricas_clean, caracterizaciones_clean)]

#print (rubricas_clean)
#print (caracterizaciones_clean)
#print (map_rubric_characterization)

df_resultados = pd.read_excel("resultados.xlsx")  


for col, wmap  in zip (df_resultados.columns[3:], map_rubric_characterization):
    print ("Iteration Started")
    print (col, wmap)
    df_resultados[col] = df_resultados[col].replace(wmap, regex=False)
    print ("Iteration ended")

#df.to_csv('output.csv', index=False) 
#print (df_resultados)
df_resultados = df_resultados[df_resultados.columns[3:]]
#transposed_df = df_resultados.T

#print (df_resultados)

df_resultados.to_csv('output.csv', index=False, encoding="utf-8")
#import ace_tools as tools; tools.display_dataframe_to_user(name="Counts of Unique Values", dataframe=counts)
```
